```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/91862/Desktop/DATASCIENCECODES//Iris.csv")
```

```python
column_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']
```

```python
print("\nFeatures and their types:")
print(df.dtypes)
```

```python
# Create histograms for each feature in the dataset
plt.figure(figsize=(12, 8))
for i, feature in enumerate(column_names[:-1]):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Create box plots for each feature in the dataset
plt.figure(figsize=(12, 8))
for i, feature in enumerate(column_names[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'Box plot of {feature}')
    plt.xlabel('Species')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```
