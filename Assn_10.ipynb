Download    the    Iris    flower    dataset    or    any    other    dataset    into    a    DataFrame.    (e.g., https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and give the inference as:
1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a box plot for each feature in the dataset.
4. Compare distributions and identify outliers.


```python
from seaborn import load_dataset

df = load_dataset('iris')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width','species']
```

```python
print("\nFeatures and their types:")
df.dtypes
```

```python
# Create histograms for each feature in the dataset
plt.figure(figsize=(12, 8))
for i, feature in enumerate(column_names[:-1]):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Create box plots for each feature in the dataset
plt.figure(figsize=(12, 8))
for i, feature in enumerate(column_names[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=feature, data=df)
    plt.title(f'Box plot of {feature}')
    plt.xlabel('Species')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```
